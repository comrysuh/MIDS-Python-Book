```python
#Exercise 6.1 initial function definition
import sympy as sp

def Jacobian(Func, Vars):
    if not isinstance(Func, sp.Matrix):
        return "Error: F must be a SymPy matrix."
    if not isinstance(Vars, list):
        return "Error: V must be a list."
    return sp.Matrix([[sp.diff(Func[i], var) for var in Vars] for i in range(len(Func))])

```

```python
#Exercise 6.1 part (1)
import numpy as np
import sympy as sp
x, y = sp.symbols('x y')
variables = [x, y]
F = np.array([x**3*sp.sin(y), y**2*sp.cos(2*x)])
print('F is:', F)

J = Jacobian(F, variables)
print("Jacobian matrix:\n",J)
```

```python
#Exercise 6.3
def Hessian(func, Vars):
    return sp.Matrix([[sp.diff(sp.diff(func, var1), var2) for var1 in Vars] for var2 in Vars])

def extreme_values(func, Vars):
    grad_f = [sp.diff(func, var) for var in Vars]  
    critical_points = sp.solve(grad_f, (Vars[0], Vars[1]), dict=True)

    H = Hessian(func, Vars)

    results = []
    for point in critical_points:
        x_val = point[Vars[0]]
        y_val = point[Vars[1]]
        f_val = func.subs(point)
        H_at_point = H.subs(point)
        det_H_at_point = H_at_point.det()
        A = H_at_point[0, 0]
    
        if det_H_at_point > 0 and A > 0:
            nature = "Minimum"
        elif det_H_at_point > 0 and A < 0:
            nature = "Maximum"
        elif det_H_at_point < 0:
            nature = "Saddle point"
        else:
            nature = "Indeterminate"
    
        results.append((x_val, y_val, f_val, nature))
     
    return results 

import sympy as sp

x, y = sp.symbols('x y')
f1 = 2*x**3+2*y**3-3*x**2+3*y**2-12*x-12*y
print('Function is',f1)
variables = [x, y]

results = extreme_values(f1, variables)

for x_val, y_val, f_val, nature in results:
    print(f"Critical Point: ({x_val}, {y_val}), Function Value: {f_val}, Nature: {nature}")
    
f2 = 2*x**3+ 2*y**3+3*y**2-9*x**2-36*y+4
print('\nFunction is',f2)
variables = [x, y]

results = extreme_values(f2, variables)

for x_val, y_val, f_val, nature in results:
    print(f"Critical Point: ({x_val}, {y_val}), Function Value: {f_val}, Nature: {nature}")
    
```

```python
#Exercise 6.4 part (1)
def Hessian(func, Vars):
    return sp.Matrix([[sp.diff(sp.diff(func, var1), var2) for var1 in Vars] for var2 in Vars])

def extreme_values_eigH(func, Vars):
    grad_f = [sp.diff(func, var) for var in Vars]  
    critical_points = sp.solve(grad_f, (Vars[0], Vars[1]), dict=True)

    H = Hessian(func, Vars)

    results = []
    for point in critical_points:
        x_val = point[Vars[0]]
        y_val = point[Vars[1]]
        f_val = func.subs(point)
        H_at_point = H.subs(point)
        eigvals_H_at_point = H_at_point.eigenvals()
    
        eigenvalues_list = list(eigvals_H_at_point.keys())
        if all(eigval > 0 for eigval in eigenvalues_list):
            nature = "Minimum"
        elif all(eigval < 0 for eigval in eigenvalues_list):
            nature = "Maximum"
        elif any(eigval > 0 for eigval in eigenvalues_list) and any(eigval < 0 for eigval in eigenvalues_list):
            nature = "Saddle point"
        else:
            nature = "Indeterminate"
    
        results.append((x_val, y_val, f_val, nature))
     
    return results 
```

```python
#Exercise 6.5 parts (1) and (2)
def Lagrange(func, constr, var1, var2, lm):
    L = func + lm * constr

    L_x = sp.diff(L, var1)
    L_y = sp.diff(L, var2)
    L_lm = sp.diff(L, lm)

    equations = [L_x, L_y, L_lm]
    solutions = sp.solve(equations, (var1, var2, lm), dict=True)
    
    return solutions

import sympy as sp

x, y, lm = sp.symbols('x y lm')

f1 = x**2 + y**2 + 1
print('For function: ',f1)
g = x - y + 1

solutions = Lagrange(f1, g, x, y, lm)

extrema = [(sol[x], sol[y], sol[lm], f1.subs({x: sol[x], y: sol[y]})) for sol in solutions]

if len(extrema)==0:
    print('The extrema is empty.')
else:    
    for x_val, y_val, lm_val, f_val  in extrema:
        print(f"Critical Point: ({x_val}, {y_val}), Function Value: {f_val}.\n The Lagrange multiplier value: {lm_val}.")

f2 = x**2 - y**2 + 1
print('\nFor function: ',f2)
g = x - y + 1

solutions = Lagrange(f2, g, x, y, lm)

extrema = [(sol[x], sol[y], sol[lm], f2.subs({x: sol[x], y: sol[y]})) for sol in solutions]

if len(extrema)==0:
    print('The extrema is empty.')
else:    
    for x_val, y_val, lm_val, f_val  in extrema:
        print(f"Critical Point: ({x_val}, {y_val}), Function Value: {f_val}.\n The Lagrange multiplier value: {lm_val}.")
```

```python
#Exercise 6.6 part (1)
def grad_ascent(inits, f, Vars, iters, epsilon, tolerance):
    x_old = inits[0]
    y_old = inits[1]
    
    def_f_x = sp.diff(f, Vars[0])
    def_f_y = sp.diff(f, Vars[1])
    
    x_values = [x_old]
    y_values = [y_old]
    
    gradient_xs =[]
    gradient_ys =[]
    
    for i in range(iters):
        grad_value_x = def_f_x.evalf(subs={Vars[0]: x_old})
        grad_value_y = def_f_y.evalf(subs={Vars[1]: y_old})
    
        x_new = x_old + epsilon*grad_value_x.subs({Vars[0]:x_old}).evalf()
        y_new = y_old + epsilon*grad_value_y.subs({Vars[1]:y_old}).evalf()
        
        x_values.append(x_new)
        y_values.append(y_new)
        
        gradient_xs.append(grad_value_x)
        gradient_ys.append(grad_value_y)
        
        print(f"Iteration {i+1}: x = {round(x_new,2):.2f}, y = {round(y_new,2):.2f}, f(x) = {round(f.evalf(subs={Vars[0]: x_new, Vars[1]: y_new}),2):.2f}") 
        
        if abs(x_new - x_old) < tolerance and abs(y_new - y_old) < tolerance:
            break
            
        x_old = x_new
        y_old = y_new
            
    return x_values, y_values, gradient_xs, gradient_ys
```

```python
#Exercise 6.6 part (2)
import sympy as sp

x, y = sp.symbols('x y')
variables =[x, y]
f = -x**2 - y**2 + 4*x+ 2*y
x_old = 0
y_old = 0
inits =[x_old, y_old]

max_iters = 100
epsilon = 0.08
tolerance = 1e-4

x_values, y_values, gradient_xs, gradient_ys = grad_ascent(inits, f, variables, max_iters, epsilon, tolerance)

print(f"Maximum value of f(x, y) = {f.evalf(subs={x: x_values[-1], y: y_values[-1]})} at x = {x_values[-1]}, y = {y_values[-1]}")
```

```python
#Exercise 6.7 part (1) - finding integral of x^2ydxdy
import sympy as sp

x, y = sp.symbols('x y')
f = x**2*y

x_lower, x_upper = -1, 1
y_lower, y_upper = 0, sp.sqrt(1-x**2)

inner_integral = sp.integrate(f, (y, y_lower, y_upper))
print(f"Inner integral with respect to y:{inner_integral}")

double_integral_1 = sp.integrate(inner_integral, (x, x_lower, x_upper))
print(f"Double integral: {double_integral_1}")

double_integral_2 = sp.integrate(f, (y, y_lower, y_upper), (x, x_lower, x_upper)) 
print(f"Double integral done in one step: {double_integral_2}")
```
