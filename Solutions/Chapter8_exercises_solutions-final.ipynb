```python
#Exercise 8.1 parts (1) and (2)
def slr(x, y):
    if x.ndim != 1 or y.ndim != 1:
        raise ValueError('Both x and y must be 1D arrays.')
    
    mean_x = np.mean(x)
    print('The average x is:', mean_x)
    mean_y = np.mean(y)
    print('The average y is:', mean_y)
    
    dot_product_1 = np.matmul((x-mean_x), (y-mean_y))
    dot_product_2 = np.matmul((x-mean_x), (x-mean_x))
    a1 = dot_product_1/dot_product_2
    a0 = mean_y-a1*mean_x
    print('The slope is:', np.round(a1,2))
    print('The intercept is:', np.round(a0,2))
    return a0, a1
```

```python
#Exercise 8.1 part (3a)
import numpy as np
import matplotlib.pyplot as plt

x_obs = np.array([[1,2,4]])
y_obs = np.array([2,4,3])

intercept, slope = slr(x_obs,y_obs)
```

```python
#Exercise 8.1 part (3b)
import numpy as np
import matplotlib.pyplot as plt

x_obs = np.array([1,2,3])
y_obs = np.array([2,4,3])

intercept, slope = slr(x_obs,y_obs)

x = np.linspace(-2, 4, 400) 
y = slope * x + intercept

plt.figure(figsize=(8, 6))
plt.scatter(x_obs, y_obs, c='red', marker='x')
plt.plot(x, y, label=f'y = {slope}x + {intercept}')


plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.grid(True)
plt.savefig('line.png')
```

```python
#Exercise 8.2 part (1)
def r2(y, y_est):
    mean_y = np.mean(y)
    coeff_det  = 1 - np.dot((y-y_est),(y-y_est))/ np.dot((y-mean_y),(y-mean_y))
    
    return coeff_det
```

```python
#Exercise 8.2 parts (2) and (3)
import numpy as np

def slr(x, y):    
    if x.ndim != 1 or y.ndim != 1:
        raise ValueError("Both x and y must be 1D arrays.")
    
    mean_x = np.mean(x)
    print('The average x is:', mean_x)
    mean_y = np.mean(y)
    print('The average y is:', mean_y)
    
    dot_product_1 = np.matmul((x-mean_x), (y-mean_y))
    dot_product_2 = np.matmul((x-mean_x), (x-mean_x))
    a1 = dot_product_1/dot_product_2
    a0 = mean_y-a1*mean_x
    print('The slope is:', np.round(a1,2))
    print('The intercept is:', np.round(a0,2))
    return a0, a1

x_obs = np.array([1,2,3])
y_obs = np.array([2,4,3])

intercept, slope = slr(x_obs,y_obs)
y_est = intercept + slope *x_obs

R2 = r2(y_obs, y_est)
print('The coefficient of determination is:', R2)

residuals = np.sum(y_obs - y_est)
print('The sum of residuals is:', residuals)

```

```python
#Exercise 8.3 - the required function
def lrmvar(X,y):
    a = np.matmul(np.matmul(np.linalg.inv(np.matmul(X.T,X)), X.T),y)
    return a
```

```python
#Exercise 8.3 (1)
import numpy as np

X = np.array([[1,1,2],[1,2,1],[1,2,2]])
y = np.array([1,2,-2])
a = lrmvar(X,y)
print('The X matrix is:\n', X)
print('The y array is:',y)
print('The estimated parameters are:',a)
```

```python
#Exercise 8.4 part (1) - same input as Exercise 8.3 part (2)
import numpy as np
from sklearn.linear_model import LinearRegression

x_obs = np.array([1,2,3,4,5,6,7])
print("x_obs as input:\n",x_obs)
x_obs = x_obs[:,np.newaxis]
print("x_obs as converted:\n",x_obs)
y_obs = np.array([5,4,5,3,3,1,1])

reg = LinearRegression().fit(x_obs, y_obs)
estimated_y = reg.predict(x_obs)

print('\nThe fitted model is saved in an instance of ',reg)
print('The estimated coefficient (gradient) is: ',reg.coef_)
print('The estimated intercept is: ', np.round(reg.intercept_,2))
print('The estimated target values of X as given by the \'line\' of best fit are:\n', estimated_y)
print('The coefficient of determination is: ',reg.score(x_obs, y_obs))
print('The predicted target value for x=3.5 is: ',reg.predict([[3.5]]))
```

#### Exercise 8.6

```python
#Exercise 8.6 part (1)
import numpy as np  
from sklearn import preprocessing
from sklearn.linear_model import SGDRegressor 
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

X = np.array([1,2,3,4,5,6,7]) 
X = X[:,np.newaxis]
y =np.array([5,4,5,3,3,1,1])

sgd_reg = SGDRegressor(alpha=0, learning_rate='constant', warm_start=True)
mses = []
n_iterations = 1000
tol = 0.37

for iteration in range(n_iterations): 
    for i, x in enumerate(X):
        sgd_reg.partial_fit(x[:,np.newaxis], y[i].ravel())
    est_y = sgd_reg.predict(X)
    mse = mean_squared_error(y, est_y)
    mses.append(mse)
    if  mse < tol:
        print(f"Training stopped at iteration {iteration} due to convergence")
        break


print('The estimated intercept is: ', np.round(sgd_reg.intercept_,2))
print('The estimated coefficient is: ', np.round(sgd_reg.coef_,2))
print('The coefficient of determination is: ', sgd_reg.score(X, y))
        
plt.plot(range(1, len(mses)+1), mses, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error')
plt.grid(True)
plt.savefig('LearningCurve_partial_fit.png')
```

```python
#Exercise 8.7 part (2)
import numpy as np  
from sklearn.linear_model import SGDRegressor 
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

X = np.array([1,2,3])
X = X[:,np.newaxis]
y =np.array([4,2,1.5])

sgd_reg_1 = SGDRegressor(max_iter=1, alpha=0, learning_rate='constant', warm_start=True)

init_intercept = 5
init_coef = 1
n_iterations = 1000
mses = []

sgd_reg_1.fit(X, y, intercept_init=init_intercept, coef_init=init_coef)
est_y = sgd_reg_1.predict(X)
mse = mean_squared_error(y, est_y)
mses.append(mse)
tol = 0.19

for i in range(2, n_iterations + 1):
    sgd_reg_1.fit(X, y)
    est_y = sgd_reg_1.predict(X)
    mse = mean_squared_error(y, est_y)
    mses.append(mse)
    if  mse < tol:
        print(f"Training stopped at iteration {i} due to convergence")
        break

estimated_y = sgd_reg_1.predict(X)
print('The estimated intercept is: ', np.round(sgd_reg_1.intercept_,2))
print('The estimated coefficient is: ',sgd_reg_1.coef_)
print('The coefficient of determination is: ',sgd_reg_1.score(X, y))

plt.plot(range(1, len(mses)+1), mses, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error')
plt.grid(True)
plt.savefig('LearningCurve_fit.png')
```
