```python
#Exercise 2.2
from sympy import FiniteSet

u = FiniteSet(1,3,6)
v = FiniteSet(1,3)
w = FiniteSet(1, 3, 6, 8)
Pu = u.powerset()
Pv = v.powerset()
Pw = w.powerset()

print('The statement that u is a proper subset of v is {}.'.format(u.is_proper_subset(v)))
print('The statement that v is a proper subset of w is {}.'.format(v.is_proper_subset(w)))
print('The powerset of u is: \n', Pu)
print("The cardinality of the powerset of u is {}. ".format(len(u.powerset())))
print('The powerset of v is: \n', Pv)
print("The cardinality of the powerset of v is {}. ".format(len(v.powerset())))
print('The powerset of w is: \n', Pw)
print("The cardinality of the powerset of w is {}. ".format(len(w.powerset())))
```

```python
#Exercise 2.3
from sympy import FiniteSet, Union, Intersection, Complement

I = FiniteSet(*range(50)) 

A = FiniteSet(0, 5, 15, 25, 35, 40)
B = FiniteSet(2,6,18,36, 40,41,42,49)
C = FiniteSet(3,5,17,36,40,48)

A_U_B = Union(A,B)
C_intersect_B = Intersection(C,B)
print(A_U_B)
print(C_intersect_B )
print(Complement(I,A_U_B))
print(A-C_intersect_B )
print(Union(A,B,C))
print(A_U_B-C)
print(Complement(I, A_U_B-C))
```

```python
#Exercise 2.4
from sympy import ConditionSet, Symbol, S, Interval, Union, Intersection
x=Symbol('x') 
I1 = Interval(4,12)
I2 = Interval(5,15)
S=ConditionSet(x, x<12, S.Integers)

print(Intersection(I1,I2))
print(Union(I1, I2))
print(Intersection(I1, S))
print(Union(I1, S))
print(S-I1)
```

```python
#Exercise 2.6
from sympy import FiniteSet

A = FiniteSet(-1,0,2)
B = FiniteSet(1,5)

C= A*B
print('The Cartesian product C is: \n', C)

for elem in C:
    print(elem)
    
D = FiniteSet((2,5),(3,4))

for elem in (C.union(D)):
    print('The element in C.union(D):',elem)
    
for elem in (C.intersect(D)):
    print('The element in C.intersect(D):',elem) 
    
for elem in (C-D):
    print('The element in C-D',elem)    
```

```python
#Exercise 2.8
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-np.pi, np.pi, 100)
y1 = 2*np.sin(x1)

y2 = np.cos(x1)/4

x2 = np.linspace(-np.pi/2.1, np.pi/2.1, 100)
y3 = np.tan(x2)-2

fig, axs = plt.subplots(1, 3)
axs[0].plot(x1, y1)
axs[0].set_title('2sin(x)')
axs[1].plot(x1, y2, 'ro')
axs[1].set_title('cos(x)/4')
axs[2].plot(x2, y3, 'g+')
axs[2].set_title('tan(x)-2')

for ax in axs.flat:
    ax.set(xlabel='Angle', ylabel='y')

for ax in axs.flat:
    ax.label_outer()
 
```

```python
#Exercise 2.9
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x1)
theta1 = np.degrees(np.arcsin(y1))

y2 = np.cos(x1)
theta2 = np.degrees(np.arccos(y2))
           
x2 = np.linspace(-np.pi/2.1, np.pi/2.1, 100)
y3 = np.tan(x2)
theta3 = np.degrees(np.arctan(y3))

fig, axs = plt.subplots(1, 3)
axs[0].plot(y1,theta1)
axs[0].set_title('arcsin(x)')
axs[1].plot(y2, theta2, 'ro')
axs[1].set_title('arccos(x)')
axs[2].plot(y3, theta3,'g+')
axs[2].set_title('arctan(x)')

for ax in axs.flat:
    ax.set(xlabel='y', ylabel='Angle in degrees')

plt.tight_layout()    
fig.savefig('arc_trig_graphs.png')    
```

```python
#Exercise 2.12
import numpy as np
import matplotlib.pyplot as plt
def compose2(f, g):
    return lambda x: f(g(x))

def external(x):
    return np.exp(x)/(np.exp(x)+1)
    
def internal(x):
    inputs = 0.5*(x + 3)
    return inputs
ext_and_int = compose2(external, internal)

x = np.linspace(-15, 10,100)
final= ext_and_int(x)
sig = external(x)

fig = plt.figure()
plt.plot(x,final)
plt.plot(x,sig, 'r')
plt.legend(['composite','sigmoid'])

```

```python
#Exercise 2.13
import numpy as np
import matplotlib.pyplot as plt

# Define the composition of two functions
def compose2(f, g): 
    return lambda x,y: f(g(x,y))

def external(x):
    return np.exp(x)/(np.exp(x)+1)

def internal(x1,x2):
    inputs = -0.5*x1 + 0.1*x2
    return inputs

ext_and_int = compose2(external, internal)

x = np.linspace(-20, 20, 1000)
y = np.linspace(-10,10,1000)
inputs = internal(x,y)
final= ext_and_int(x,y)

fig, axs = plt.subplots(1,3)

axs[0].plot(x, final)
axs[0].set_xlabel('x')
axs[1].plot(y, final, 'tab:red')
axs[1].set_xlabel('y')
axs[2].plot(inputs, final, 'tab:red')
axs[2].set_xlabel('inputs')

for ax in axs.flat:
    ax.set(ylabel='sigmoid')

for ax in axs.flat:
    ax.label_outer()
```
