```python
#Exercise 5.1
from sympy import Limit, Symbol
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.symbols('x')

expr = (sp.Abs(x)+x)/x
print(expr)
limit_right = Limit(expr, x, 0, dir='+').doit()
limit_left = sp.limit(expr, x, 0, dir='-')

print("Limit from the right:", limit_right)
print("Limit from the left:", limit_left)

def f(x):
    return np.where(np.abs(x) < 0.01, np.nan, (np.abs(x)+x)/x)

x_vals = np.linspace(-4, 4.01, 400)
f_vals = f(x_vals)

fig = plt.figure(figsize=(8, 4))
plt.plot(x_vals, f_vals, label='$\\frac{|x|+x}{x}$', color='blue')
plt.scatter([0], [limit_right], color='red', zorder=5, label='Limit from the right')
plt.scatter([0], [limit_left], color='green', zorder=5, label='Limit from the left')
plt.title("Visualisation of f(x) near x=0")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axvline(x=0, color='gray', linestyle='--')
plt.ylim(-1, 5)
plt.legend()
plt.grid(True)

fig.savefig('Limits.eps', dpi=600)
```

```python
#Exercise 5.3 part (1)
from sympy import *

x, u = symbols('x u')
g = sin(x)
f = exp(u)

composite_function = f.subs(u, g)

dg = diff(g, x)  
df = diff(f, u).subs(u, g) 
chain_rule_result = df * dg

print(f"Function g(x) = {g}")
print(f"Function f(u) = {f}")
print(f"Composite function f(g(x)) = {composite_function}")
print(f"Derivative of g(x), g'(x), = {dg}")
print(f"Derivative of f(u) at u = g(x), f'(g(x)), = {df}")
print(f"Result of the chain rule, (f(g(x)))' = {chain_rule_result}")

f1 = exp(sin(x))
d1 = f1.diff(x)
print('The derivative of the function directly is: ', d1)
```

```python
#Exercise 5.4
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Define the symbol
x = sp.Symbol('x')

# Define the function
f = (x**3 - 3*x**2 - x + 3) / (x - 3)

# 1. Compute the limit
limit_value = sp.limit(f, x, 3)
print('1. Limit of f(x) as x approaches 3:', limit_value)

# 2. Find the derivative
derivative_f = sp.diff(f, x)
print('2. Derivative of f(x):', derivative_f)

# 3. Plot the function
x_values = np.linspace(0, 5, 400)
f_func = sp.lambdify(x, f, 'numpy')
y_values_f = f_func(x_values)

derivative_f_func = sp.lambdify(x, derivative_f, 'numpy')
y_values_derivative_f = derivative_f_func(x_values)

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values_f, label='f(x)')
plt.plot(x_values, y_values_derivative_f, label="f'(x)")

plt.xlabel('x')
plt.ylabel('Function values')
plt.title("Plot of f(x) and its gradient function f'(x)")
plt.grid(True)
plt.legend()
```

```python
#Exercise 5.5 parts (1) and (2)
import sympy as sp

x = sp.symbols('x')
expression = sp.sin(x)/x

# (1) Compute the limit
limit_value = sp.limit(expression, x, 0)

print('The limit of sin(x)/x as x approaches 0 is:', limit_value)

#(2) Visualisation 
import numpy as np
import matplotlib.pyplot as plt

func = sp.lambdify(x, expression, 'numpy')

x_vals = np.linspace(-10, 10, 400)

# We manually set y(0) = 1 since sin(0)/0 is not defined
y_vals = np.where(x_vals == 0, 1, func(x_vals))  

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r'$\frac{\sin(x)}{x}$')

# Emphasise the limit at x = 0, y = 1
plt.scatter([0], [1], color='red')  
plt.title('Plot of $\sin(x)/x$')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
```

```python
#Exercise 5.5 part (3) L’Hopital’s Rule
import sympy as sp
x = sp.symbols('x')

diff_numerator = sp.diff(sp.sin(x), x)
print('differentiation of the numerator is:',diff_numerator)
diff_denominator = sp.diff(x, x)
print('differentiation of the denominator is:',diff_denominator)

quotient = diff_numerator/diff_denominator
limit_value = sp.limit(quotient, x, 0)
print('The limit of sin(x)/x as x approaches 0 is:', limit_value)
```

```python
#Exercise 5.6
from sympy import Symbol, diff, solve
import numpy as np

x = Symbol('x')

def my_max_min(f,x):
    print('\n',f)
    ans = np.array(['is a maximum','is a minimum'])

    f_1st_diff = diff(f,x)

    critical_x_values = solve(f_1st_diff,x)
    print('There are {} roots for the first derivative function, which are {}.'.format(len(critical_x_values),critical_x_values))
    f_2nd_diff = diff(f,x,2)

    for i in critical_x_values:
        A = i
        B = f_2nd_diff.subs(x,i).evalf()
        C = f.subs(x,i).evalf()
        if B < 0:
            type = ans[0]
        elif B > 0:
            type = ans[1]
    
        print('When x value is {:.1f}, the 2nd derivative value is {:.1f},\n\t the function value is {:.2f} and it {}'.format(A, B, C,type))
    return

f = x**3 - 3*x**2 -24*x +3
my_max_min(f,x)
f = 4*x**3 - 3*x**2 + 1
my_max_min(f,x)
```

```python
#Exercise 5.7
from sympy import Symbol, diff, solve
import numpy as np


x = Symbol('x')


def my_max_min1(f,x):
    print('\n',f)
    ans = np.array(['is a maximum','is a minimum','is a point of inflection'])

    f_1st_diff = diff(f,x)

    critical_x_values = solve(f_1st_diff,x)
    print('There are {} roots for the first derivative function, which are {}.'.format(len(critical_x_values),critical_x_values))
    f_2nd_diff = diff(f,x,2)

    for i in critical_x_values:
        A = i
        B = f_2nd_diff.subs(x,i).evalf()
        C = f.subs(x,i).evalf()
        if B < 0:
            type = ans[0]
        elif B > 0:
            type = ans[1]
        elif B == 0:
            left = f_1st_diff.subs(x,i-0.01).evalf()
            right = f_1st_diff.subs(x,i+0.01).evalf()
            if (left > 0 and right < 0):
                type = ans[0]
            elif (left < 0 and right > 0):
                type = ans[1]
            else:
                type = ans[2]
        print('When x value is {:.1f}, the 2nd derivative value is {:.1f},\n\t the function value is {:.1f} and it {}'.format(A, B, C,type))
    return

f = x**4
my_max_min1(f,x)
f = x**3
my_max_min1(f,x)
f = 1 - x**4
my_max_min1(f,x)
```

```python
#Exercise 5.8
import sympy as sp

x, u = sp.symbols('x u')
f1 = x/(x**2-1)
print('\n',f1)
indef_integral1 = sp.integrate(f1, x) 
print('The indefinite integral is:', indef_integral1 + sp.Symbol('C'))

print('\nSuggested substitution: u = x**2-1')
du_dx = sp.diff(x**2-1, x) 
print('The differentiation of the substitution is',du_dx)

integrand_subs = (1/u) * (1/2)

integral_subs = sp.integrate(integrand_subs, u)
print('Integral with substitution:',integral_subs)
print('Original integral:',integral_subs.subs(u,x**2-1) + sp.Symbol('C'))

f2 = x**2*sp.cos(x**3)
print('\n',f2)
indef_integral2 = sp.integrate(f2, x) 
print('The indefinite integral is:', indef_integral2 + sp.Symbol('C'))

print('\nSuggested substitution: u = x**3')
du_dx = sp.diff(x**3, x) 
print('The differentiation of the substitution is',du_dx)

integrand_subs = (sp.cos(u)) * (1/3)

integral_subs = sp.integrate(integrand_subs, u)
print('Integral with substitution:',integral_subs)
print('Original integral:',integral_subs.subs(u,x**3) + sp.Symbol('C'))
      
f3 = sp.sin(4*x)
print('\n',f3)
indef_integral3 = sp.integrate(f3, x) 
print('The indefinite integral is:', indef_integral3 + sp.Symbol('C'))
result = sp.integrate(f3, (x, 0, sp.pi/8)) 
print('The definite integral is:',result)

print('\nSuggested substitution: u = 4*x')
du_dx = sp.diff(4*x, x) 
print('The differentiation of the substitution is',du_dx)

integrand_subs = (sp.sin(u)) * (1/4)

integral_subs = sp.integrate(integrand_subs, u)
print('Integral with substitution:',integral_subs)
print('Original integral:',integral_subs.subs(u,4*x) + sp.Symbol('C'))

lower_limit = 4*0        
upper_limit = 4*sp.pi/8       

integral_subs = sp.integrate(integrand_subs, (u, lower_limit, upper_limit))
print('The definite integral is:',integral_subs)

f4 = x/(1+x)**3
print('\n',f4)
indef_integral4 = sp.integrate(f4, x) 
print('The indefinite integral is:', indef_integral4 + sp.Symbol('C'))
result = sp.integrate(f4, (x, 0, 1)) 
print('The definite integral is:',result)

print('\nSuggested substitution: u = 1+x')
du_dx = sp.diff(x+1, x) 
print('The differentiation of the substitution is',du_dx)

integrand_subs = ((u-1)/u**3) * (1)

integral_subs = sp.integrate(integrand_subs, u)
print('Integral with substitution:',integral_subs)
print('Original integral:',integral_subs.subs(u,1+x) + sp.Symbol('C'))

lower_limit = 1+0        
upper_limit = 1+1       

integral_subs = sp.integrate(integrand_subs, (u, lower_limit, upper_limit))
print('The definite integral is:',integral_subs)
```

```python
#Exercise 5.9
import sympy as sp
import numpy as np

x = sp.symbols('x')
f1 = x*sp.log(x)
print('\nFunction1 is:',f1)
indef_integral1 = sp.integrate(f1, x) 
print('The indefinite integral is', indef_integral1 + sp.Symbol('C'))

u = sp.log(x)  
dv = x  

du = sp.diff(u, x)
v = sp.integrate(dv, x)

integral = u * v - sp.integrate(v * du, x) + sp.Symbol('C')

print(f'By parts:\nu = {u}')
print(f'dv = {dv}')
print(f'du = {du}')
print(f'v = {v}')
print('Integral by parts:', integral)

f2 = x*sp.exp(x)
print('\nFunction2 is:',f2)
indef_integral2 = sp.integrate(f2, x) 
print('The indefinite integral is', indef_integral2 + sp.Symbol('C'))

u = x  
dv = sp.exp(x)

du = sp.diff(u, x)
v = sp.integrate(dv, x)

integral = u * v - sp.integrate(v * du, x) + sp.Symbol('C')

print(f'By parts:\nu = {u}')
print(f'dv = {dv}')
print(f'du = {du}')
print(f'v = {v}')
print('Integral by parts:', integral)

f3 = x**2*sp.cos(x)
print('\nFunction3 is:',f3)
indef_integral3 = sp.integrate(f3, x) 
print('The indefinite integral is', indef_integral3 + sp.Symbol('C'))
result = sp.integrate(f3, (x, 0, sp.pi/2)) 
print('The definite integral is:',result)

u = x**2  
dv = sp.cos(x)

du = sp.diff(u, x)
v = sp.integrate(dv, x)

integral = u * v - sp.integrate(v * du, x) + sp.Symbol('C')

print(f'By parts:\nu = {u}')
print(f'dv = {dv}')
print(f'du = {du}')
print(f'v = {v}')

print('\nStill have a complicated integral of 2x*Sin(x) so use parts again:')

u1 = 2*x  
dv1 = sp.sin(x)

du1 = sp.diff(u1, x)
v1 = sp.integrate(dv1, x)

integral = u * v - u1*v1 + sp.integrate(v1 * du1, x) + sp.Symbol('C')

print(f'u1 = {u1}')
print(f'dv1 = {dv1}')
print(f'du1 = {du1}')
print(f'v1 = {v1}')
print('Final integral by parts:', integral)

f = sp.lambdify(x,integral,"numpy")
print('Indefinite integral is:',f(np.pi/2)-f(0))

f4 = sp.log(x)
print('\nFunction4 is:',f4)
indef_integral4 = sp.integrate(f4, x) 
print('The indefinite integral is', indef_integral4 + sp.Symbol('C'))
result = sp.integrate(f4, (x, 1, 2)) 
print('The definite integral is:',result)

u = sp.log(x)  
dv = 1

du = sp.diff(u, x)
v = sp.integrate(dv, x)

integral = u * v - sp.integrate(v * du, x) + sp.Symbol('C')

print(f'By parts:\nu = {u}')
print(f'dv = {dv}')
print(f'du = {du}')
print(f'v = {v}')
print('Integral by parts:', integral)

f = sp.lambdify(x,integral,"numpy")
print('Indefinite integral is:',f(2)-f(1))
```

```python
#Exercise 5.10
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

x = sp.symbols('x')
g = x*sp.exp(x**2)

# Compute the indefinite integral
indefinite_integral = sp.integrate(g, x) 
print(f"The indefinite integral of g(x) is: {indefinite_integral}")


definite_integral = sp.integrate(g, (x, -2, 2))
print(f"The definite integral of g(x) from -2 to 2 is: {definite_integral}")

# Plot the original function
f_lambdified = sp.lambdify(x, g, 'numpy')
integral_lambdified = sp.lambdify(x, indefinite_integral, 'numpy')

x_vals = np.linspace(-2, 2, 400)
f_vals = f_lambdified(x_vals)
integral_vals = np.array([integral_lambdified(val) for val in x_vals])

plt.plot(x_vals, f_vals, label='g(x) = xe^{x^2}')
plt.plot(x_vals, integral_vals, label='Integral of g(x)', linestyle='--')
plt.legend()
plt.title('Original Function and Its Indefinite Integral')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

# Plot the definite integral area
x_vals_filled = np.linspace(-2, 2, 200)
y_vals_filled = f_lambdified(x_vals_filled)

plt.plot(x_vals, f_vals, label='g(x) = xe^{x^2}')
plt.fill_between(x_vals_filled, y_vals_filled, alpha=0.3, color='orange', label='Definite Integral Area')
plt.legend()
plt.title('Definite Integral of g(x) from -2 to 2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

int1_g = sp.integrate(g, (x, 0, 2))
int2_g = sp.integrate(g,(x, -2, 0))
total1 = int1_g.evalf()
total2 = int2_g.evalf()
print('Area under the curve of g(x) from 0 to 2:',total1)
print('Area under the curve of g(x) from -2 to 0:',total2)
print('Real area between the curve of g(x) and the x-axis from  -2 to 2 is:\n',total1 - total2)
```
